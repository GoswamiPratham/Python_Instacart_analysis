```python
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

# Question 4

```python
# creat df with counts of order/hour
```

```python
counts = df_ords_merged['hour_ordered'].value_counts().rename_axis('hour').reset_index(name='counts')
```

```python
counts
```

```python
orders = list(df_ords_merged['hour_ordered'].value_counts().quantile([0.25, 0.5, 0.75]))
```

```python
orders
```

```python
# creating list
```

```python
few = list(counts.loc[counts['counts'] <= orders[0]]['hour'])
```

```python
average = list(counts[(counts['counts'] > orders[0]) & (counts['counts'] < orders[2])]['hour'])
```

```python
most = list(counts.loc[counts['counts'] >= orders[2]]['hour'])
```

```python
# Creating index
```

```python
index_few = df_ords_merged['hour_ordered'].isin(few)
```

```python
index_average = df_ords_merged['hour_ordered'].isin(average)
```

```python
index_most = df_ords_merged['hour_ordered'].isin(most)
```

```python
# Creating new collums and labels
```

```python
df_ords_merged.loc[index_few,'busiest_period_of_day'] = 'Fewest orders'
```

```python
df_ords_merged.loc[index_average,'busiest_period_of_day'] = 'Average orders'
```

```python
df_ords_merged.loc[index_most,'busiest_period_of_day'] = 'Most orders'
```
